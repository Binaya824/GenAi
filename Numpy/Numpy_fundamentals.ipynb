# Numpy Fundamentals

### Numpy is the fundamental package for scientific computing in python. It is a python library that provides a Multi dimensional Array object and various derived objects Masked Arrays or Matrices and an assortment of routine for fast operations on Arrays including Mathematical , Logical , Shape Manipulation , Sorting , Selecting , I/O , Discrete Fourier Transform , Basic Liner Algebra , Statistical Operation , Random Simulation , etc.

### At the core the Numpy Package is the n-dimensional Array Object.

## Numpy Arrays VS Python Sequences

### Numpy Arrays have a fixed size at creation , unlike Python Lists which grow dynamically. Changing the size of an nd array will create the new array and delete the origin al array.

### The element of the Numpy Array are required to be the same data type and thus will be the same size in the memory.

### Numpy Arrays facilitate advanced Mathematical and other types of Operations on large number of Data . Typically , such operations are executed more efficiently with less code than is possible using Python's built in sequences.

# Attributes

```python
array = np.array([1,2,3,4,5])
print(array.ndim)
print(array.shape)
print(array.size)
print(array.itemsize)
```

```python
array = np.array([[[1,2],[3,4]],[[5,6],[7,8]]] , dtype=np.int32)
print(array.ndim)
print(array.shape)
print(array.size)
print(array.itemsize)
print(array.dtype)
```

```python
arr = np.arange(1,10).reshape(3,3)
print(arr)
arr = arr.astype('float64')
print(arr)
```
